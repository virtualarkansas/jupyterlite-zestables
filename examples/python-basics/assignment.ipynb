## Exercise 1: Variables and Types

Create variables of different types and print them.

```python
# Create a string variable with your name
name = "Student"

# Create an integer variable with your age
age = 15

# Print both variables
print(f"Hello, my name is {name} and I am {age} years old.")
print(f"name is type: {type(name).__name__}")
print(f"age is type: {type(age).__name__}")
```

## Exercise 2: Lists and Loops

Create a list and use a for loop to process it.

```python
# Create a list of your favorite colors
colors = ["blue", "green", "red", "purple"]

# Use a for loop to print each color
for i, color in enumerate(colors, 1):
    print(f"{i}. {color}")

print(f"\nTotal colors: {len(colors)}")
```

## Exercise 3: Functions

Write a function that calculates the area of a rectangle.

```python
def rectangle_area(width, height):
    """Calculate the area of a rectangle."""
    return width * height

# Test your function
print(f"Area of 5x3 rectangle: {rectangle_area(5, 3)}")
print(f"Area of 10x7 rectangle: {rectangle_area(10, 7)}")
print(f"Area of 2.5x4.2 rectangle: {rectangle_area(2.5, 4.2)}")
```
