# Data Analysis with Python

Learn to analyze data using Python. We'll work with lists and basic statistics

## Part 1: Working with Data

Here's a dataset of daily temperatures (in Fahrenheit) for two weeks.

```python
# Daily high temperatures for 14 days
temperatures = [72, 75, 68, 71, 77, 80, 82, 79, 74, 70, 68, 73, 76, 78]
days = list(range(1, len(temperatures) + 1))

print("Day | Temperature")
print("-" * 20)
for day, temp in zip(days, temperatures):
    print(f"  {day:2d} | {temp} F")
```

## Part 2: Basic Statistics

Calculate the mean, minimum, maximum, and range of the temperatures.

```python
mean_temp = sum(temperatures) / len(temperatures)
min_temp = min(temperatures)
max_temp = max(temperatures)
temp_range = max_temp - min_temp

print(f"Mean temperature:  {mean_temp:.1f} F")
print(f"Min temperature:   {min_temp} F")
print(f"Max temperature:   {max_temp} F")
print(f"Temperature range: {temp_range} F")
```

## Part 3: Text Visualization

Create a simple bar chart using text characters.

```python
print("Temperature Bar Chart")
print("=" * 40)

for day, temp in zip(days, temperatures):
    bar_length = temp - 60  # Offset so bars aren't too long
    bar = '#' * bar_length
    print(f"Day {day:2d} | {bar} {temp}F")
```
